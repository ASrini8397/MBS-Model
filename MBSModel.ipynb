```python
import matplotlib.pyplot as plt
```

```python
def MBSModel(CurrentBalance, PTRate, WAC, WAM, PSA, CPR):
    intarray=[]
    principalarray=[]
    paymentarray=[]
    MonthlyWac=WAC/12
    MonthlyPT= PTRate/12
    totalcashflow=0
    outstandingprincipal=CurrentBalance
    for i in range(0, WAM):
        if (i< 30):
            SMM= 1-(1-CPR*((i+6)/30)*PSA)**.083
       
        SMM= 1-(1-CPR*PSA)**.083
        mp= CurrentBalance*(((MonthlyWac*(MonthlyWac+1)**WAM))/(((1+MonthlyWac)**WAM)-1))
#         print(mp)
        interest= outstandingprincipal*MonthlyWac
        intarray.append(interest)
#         print("interest", interest)
        netInt= interest*MonthlyPT
        principal= mp-interest
        prepayment= SMM*(CurrentBalance-principal)
        totalprincipal=  principal+ prepayment
        principalarray.append(totalprincipal)
        paymentarray.append(mp)
#         print("principal" ,totalprincipal)
        if(outstandingprincipal- totalprincipal <=0):
            cashflow+= outstandingprincipal
            outstandingprincipal=0
            break
        else:
            cashflow = netInt+ totalprincipal
            totalcashflow+= cashflow

            outstandingprincipal-= totalprincipal
    #         print("Principal left", outstandingprincipal)
    #         print(WAM)
            if (i%12==0 and i>0):
                print ("Yearly Cashflows:", totalcashflow)
            x_data1=intarray
            x_data2=principalarray
#             x_data3= mp
            plt.plot(x_data1)
            plt.plot(x_data2)
            plt.plot(paymentarray)
#             plt.plot(x_data3)
            plt.xlabel("Time in Months")
            plt.ylabel("Break down of Monthy Payments")
    return totalcashflow
        
        
        
        
   
```
